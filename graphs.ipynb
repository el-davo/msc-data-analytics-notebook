```python
import matplotlib.pyplot as plt
import seaborn as sns

fig1 = plt.figure(figsize=(10, 10))


x_values = [1,2,3,4]
y_values = [1,2,3,4]

y_error = [1,0.5,0.75,0.25]

plt.errorbar(x_values, y_values,  yerr=y_error, fmt='o', markersize=2, capsize=5)
plt.plot([0, 5], [0, 5], linewidth=3)

plt.title("Measuring Model Error")
plt.xlabel('X')
plt.ylabel('Y')

# plt.text(0.1, 1.5, "Y\u2082".format(1))

plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

fig1 = plt.figure(figsize=(10, 10))
np.random.seed(sum(map(ord, "distributions")))

x = np.random.normal(size=300)
sns.distplot(x);

plt.plot([0, 0], [0, 0.445], linewidth=2, linestyle='--')

plt.title("Dispersion of errors")
plt.xlabel('Values')
plt.ylabel('Occurences')
plt.text(-0.5, 0.10,"Central Tendancy", ha='left', va='top')
```
